```python
from shapely.geometry import Point, LineString, Polygon
import matplotlib.pyplot as plt

point = Point(1, 1)
```

```python
# Lista de puntos que definen la línea
points = [(0, 0), (0.5,0.5),(1, 0),(0,0), (0, -0.6), (1,-0.6), (1, 0)]

# Crear la línea utilizando la lista de puntos
line = LineString(points)

# Extraer las coordenadas x e y de la línea
x, y = line.xy

# Crear el gráfico
fig, ax = plt.subplots()
ax.plot(x, y)

# Configurar el aspecto del gráfico
ax.set_aspect('equal')
ax.grid(True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gráfico de una línea Shapely')

# Mostrar el gráfico
plt.show()
```

```python
# Crear puntos
point1 = Point(0, 0)
point2 = Point(3, 1)
point3 = Point(0, 3)

# Crear línea
line1 = LineString([point1, point2, point3])

# Crear una figura
fig, ax = plt.subplots()

# Plotear los puntos
plt.plot(*point1.xy, 'ro')  # Punto 1 en rojo
plt.plot(*point2.xy, 'ro')  # Punto 2 en rojo
plt.plot(*point3.xy, 'ro')  # Punto 3 en rojo

# Plotear la línea
plt.plot(*line1.xy)

# Establecer límites del gráfico
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 4])

# Mostrar el gráfico
plt.show()
```

```python
point1 = Point(0, 0)
point2 = Point(3, 1)
point3 = Point(0, 3)

# Crear el polígono
polygon2 = Polygon([[p.x, p.y] for p in [point1, point2, point3]])

# Extraer las coordenadas del polígono
x, y = polygon2.exterior.xy

# Graficar el polígono
plt.plot(x, y, color='blue')
plt.fill(x, y, color='lightblue', alpha=0.5)  # Rellenar el polígono
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

```python

# Definir el polígono exterior y el agujero
exterior = [(0, 0), (3, 1), (0, 3)]
hole = [[(0, 1), (3, 1), (0, 1.5)]]

# Crear el polígono con agujero
polygon_with_hole = Polygon(shell=exterior, holes=hole)

# Extraer las coordenadas del polígono exterior y el agujero
exterior_x, exterior_y = zip(*exterior)
hole_x, hole_y = zip(*hole[0])

# Graficar el polígono con agujero
plt.figure()
plt.plot(exterior_x, exterior_y, 'b-')  # Polígono exterior en azul
plt.plot(hole_x, hole_y, 'r-')          # Agujero en rojo
plt.fill(exterior_x, exterior_y, color='blue', alpha=0.5)  # Rellenar el polígono exterior
plt.fill(hole_x, hole_y, color='white')  # No rellenar el agujero
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono con agujero')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

```python
# Definir el polígono exterior
exterior_coords = [(0, 0), (5, 0), (5, 5), (0, 5)]
exterior_polygon = Polygon(exterior_coords)

# Definir el polígono interior
interior_coords = [(0.3, 0.3), (4.7, 0.3), (4.7, 4.7), (0.3, 4.7)]
interior_polygon = Polygon(interior_coords)
# Definir un polígono rojo entre el exterior y el interior
offset = 0.1  # Ajuste para el tamaño del polígono rojo
red_polygon_coords = [(x + offset, y + offset) for x, y in interior_coords]
red_polygon = Polygon(red_polygon_coords)

# Crear el polígono con el agujero
polygon_with_hole = exterior_polygon.difference(interior_polygon)

# Extraer las coordenadas del polígono con agujero
x_exterior, y_exterior = exterior_polygon.exterior.xy
x_interior, y_interior = interior_polygon.exterior.xy
x_with_hole, y_with_hole = polygon_with_hole.exterior.xy
x_red, y_red = red_polygon.exterior.xy

# Graficar los polígonos
plt.figure(figsize=(8, 8))

# Polígono exterior
plt.plot(x_exterior, y_exterior, color='black')
plt.fill(x_exterior, y_exterior, color='lightblue', alpha=0.5)

# Polígono interior (rellenado de blanco)
plt.plot(x_interior, y_interior, color='black', alpha=0.5)
plt.fill(x_interior, y_interior, color='white')



plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono con agujero')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid(True)
plt.show()
```
